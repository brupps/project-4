```python
# import pandas and load and preview the dataset
import pandas as pd
df = pd.read_csv("Resources/movies_cleaned.csv")
df.head()
```

```python
# display columns
df.columns
```

```python
# check data types
df.dtypes
```

```python
# add new column that calculates the net
df["net"] = df["gross"] - df["budget"]
df.head()
```

```python
# show data types again
df.dtypes
```

```python
# create column for the percentage the gross is of the budget
df["percentage"] = ((df["gross"]/df["budget"]) * 100)
df.head()
```

```python
import numpy as np

bins = [0, 100, 200, 300]
labels = np.array([1, 2, 3, 4])
df["profit_category"] = labels[np.searchsorted(bins, df["percentage"])-1]

df.head()
```

```python
# show datatypes
df.dtypes
```

```python
# check for outliers in the budget column
import matplotlib.pyplot as plt

budgets = df["budget"]
fig1, ax1 = plt.subplots()
ax1.set_title('Budget Column')
ax1.boxplot(budgets)
plt.show()
```

```python
# Determine which data points are outside of the 1.5 IQR range
quartiles = np.quantile(budgets,[.25,.75])
iqr = quartiles[1]-quartiles[0]
lower_bound = quartiles[0]-(1.5*iqr)
upper_bound = quartiles[1]+(1.5*iqr)


print(lower_bound)
print(upper_bound)
```

```python
# check for outliers in the percentage column
import matplotlib.pyplot as plt

percents = df["percentage"]
fig1, ax1 = plt.subplots()
ax1.set_title('Percentage Column')
ax1.boxplot(percents)
plt.show()
```

```python
# Determine which data points are outside of the 1.5 IQR range
quartiles = np.quantile(percents,[.25,.75])
iqr = quartiles[1]-quartiles[0]
lower_bound = quartiles[0]-(1.5*iqr)
upper_bound = quartiles[1]+(1.5*iqr)


print(lower_bound)
print(upper_bound)
```

```python
# create a new df containing only the numerical columns so we can look for correlations
#first make a copy

num_df = df.copy()
num_df.head()
```

```python
# drop categorical/string columns
num_df = num_df.drop(columns=['name', 'rating', 'genre', 'released', 'director', 'writer', 'star', 'country', 'company'])

num_df.head()
```

```python
# use correlation to compare columns to find relationships between numerical features; show as heatmap

import seaborn as sns
%matplotlib inline

sns.heatmap(num_df.corr(), cmap='YlGnBu', annot=True, linewidths = 0.2)
```

```python
# create scatterplot for strongest correlation pair: budget vs gross

import hvplot.pandas

budget_gross_plot = num_df.hvplot.scatter(
    x="budget",
    y="gross",
    title="Expected Gross Based on Budget"
)
budget_gross_plot
```

```python
# prep data for linear regression

X = num_df["budget"].values.reshape(-1,1)
y = num_df["gross"]
```

```python
# build linear regression model

from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')

model = LinearRegression()
model.fit(X, y)

# display model's best fit formula

print(f"Model's formula: y = {model.intercept_} + {model.coef_[0]}X")
```

```python
# make predictions using the X set
predicted_gross = model.predict(X)
```

```python
# make new dataframe with a new column for predictions

pred_gross_df = num_df.copy()

pred_gross_df["pred_gross"] = predicted_gross

pred_gross_df.head()
```

```python
# plot line of best fit & superimpose it over the scatter plot
best_fit_line = pred_gross_df.hvplot.line(
    x = "budget",
    y = "pred_gross",
    color = "red"
)

budget_gross_plot * best_fit_line
```

```python
# Assess model

from sklearn.metrics import mean_squared_error, r2_score

score = model.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_gross)
mse = mean_squared_error(y, predicted_gross)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```

```python
# repeat process for profit_category since that's our agreed upon target?

#scatterplot

profit_plot = num_df.hvplot.scatter(
    x="budget",
    y="profit_category",
    title="Expected Profit Category Based on Budget"
)
profit_plot
```

```python
# set up training data set to do multi-feature regression & random forest

df_train = num_df.copy()

df_train.head()

```

```python
# because profit_category is a binning of percentage, remove that from the set.
# net is also not needed as we determined to not use it as a target

df_train = df_train.drop(columns=["net", "percentage"])
df_train.head()
```

```python
# dependencies for scaling and random forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# define X and y for training
X = df_train.drop(columns="profit_category")
y = df_train["profit_category"]

#split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# scale the data

scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# linear regression on the training data
#set it up
lin_model = LinearRegression()

#fit it
lin_model.fit(X_train, y_train)

#predict
lin_predict = lin_model.predict(X_test)

print("R Square: ", r2_score(y_test, lin_predict))
```

```python
# linear regression on the SCALED training data
#set it up
scaled_linear = LinearRegression()

#fit it
scaled_linear.fit(X_train_scaled, y_train)

#predict
scaled_lin_predict = scaled_linear.predict(X_test_scaled)

print("R Square: ", r2_score(y_test, scaled_lin_predict))
```

```python
# random forest

#set it up
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)

#fit it
rf_model = rf_model.fit(X_train_scaled, y_train)

#predict
predictions = rf_model.predict(X_test_scaled)
```

```python
# calculate accuracy score

accuracy_score(y_test, predictions)

# NOTE: This accuracy score used data that included the gross, which is essentially what we are trying to predict.
# This means this isn't really a valid predictive model
```
