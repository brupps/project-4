```python
# import pandas and load and preview the dataset
import pandas as pd
df = pd.read_csv("Resources/movies_cleaned.csv")
df.head()
```

```python
# create column for the percentage the gross is of the budget
df["percentage"] = ((df["gross"]/df["budget"]) * 100)
df.head()
```

```python
# create categories/bins for the percentage and add to dataframe
import numpy as np

bins = [0, 200]
labels = np.array([0, 1])
df["profit_category"] = labels[np.searchsorted(bins, df["percentage"])-1]


df.head()
```

```python
# create a copy of the dataframe and drop columns that will not be used as features or the target
df_copy = df.copy()

df_copy = df_copy.drop(columns=['name', 'released', 'writer', 'gross', 'company', 'percentage',
                                'score', 'votes'])
df_copy.head()
```

```python
# drop all but US companies
us_df = df_copy[df_copy['country'].isin(['United States'])]

us_df
```

```python
# count the genres by category
genre_counts = us_df['genre'].value_counts()
genre_counts
```

```python
# put all those under 50 into an "other" category
# use a variable to create a list of replacements needed
other_cat = genre_counts[genre_counts < 50].index.tolist()

# use for loop to replace those in the dataframe
for cat in other_cat:
    us_df['genre'] = us_df['genre'].replace(cat,'Other')

# check the new counts
us_df['genre'].value_counts()
```

```python
# check ratings counts
rating_counts = us_df['rating'].value_counts()
rating_counts
```

```python
# put all those under 50 into an "other" category
# use a variable to create a list of replacements needed
other_rate = rating_counts[rating_counts < 75].index.tolist()

# use for loop to replace those in the dataframe
for rate in other_rate:
    us_df['rating'] = us_df['rating'].replace(rate,'other_rate')

# check the new counts
us_df['rating'].value_counts()
```

```python
# drop country & YEAR as we don't need it for the model
us_df = us_df.drop(columns=['country'])

us_df.head()
```

```python
# get dummies for non-numerical columns
dummies_df = pd.get_dummies(data=us_df)

dummies_df.head()
```

```python
# dependencies for logistic regression

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# define features and target
from sklearn.model_selection import train_test_split

X = dummies_df.drop(columns="profit_category")
y = dummies_df["profit_category"]

#split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# train the data
classifier.fit(X_train, y_train)
```

```python
# score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# predict outcomes
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
# calculate accuracy score
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```
