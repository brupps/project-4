```python
# import pandas and load and preview the dataset
import pandas as pd
df = pd.read_csv("Resources/movies_cleaned.csv")
df.head()
```

```python
# create column for the percentage the gross is of the budget
df["percentage"] = ((df["gross"]/df["budget"]) * 100)
df.head()
```

```python
# create categories/bins for the percentage and add to dataframe
import numpy as np

bins = [0, 100, 200, 300]
labels = np.array([1, 2, 3, 4])
df["profit_category"] = labels[np.searchsorted(bins, df["percentage"])-1]

df.head()
```

```python
# create a copy of the dataframe and drop columns that will not be used as features or the target
df_copy = df.copy()

df_copy = df_copy.drop(columns=['name', 'year', 'released', 'director', 'writer', 'star', 'company', 'percentage'])
df_copy.head()
```

```python
# use get dummies to encode the non-numerical columns
dummies_df = pd.get_dummies(data=df_copy)

dummies_df.head()
```

```python
# dependencies for scaling and random forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# define X and y for training
X = dummies_df.drop(columns="profit_category")
y = dummies_df["profit_category"]

#split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# scale the data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#set it up
rf_model = RandomForestClassifier(n_estimators=500, random_state=78)

#fit it
rf_model = rf_model.fit(X_train_scaled, y_train)

#predict
predictions = rf_model.predict(X_test_scaled)
```

```python
# calculate accuracy score

accuracy_score(y_test, predictions)
```
